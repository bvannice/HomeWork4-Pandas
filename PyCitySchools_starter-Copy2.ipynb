```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)


# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Create a dataframe to hold the above results

```python
#Total numbers
totalSchools = school_data_complete["school_name"].nunique()
totalStudents = school_data_complete["Student ID"].count()
totalBudget = school_data_complete["budget"].unique().sum()
#average math & reading
avgMath = school_data_complete["math_score"].mean()
avgReading = school_data_complete["reading_score"].mean()

#Passing counts Math, Reading & Overall
passingMath = school_data_complete[school_data_complete["math_score"] > 70].count()["school_name"]
passingReading = school_data_complete[school_data_complete["reading_score"] > 70].count()["school_name"]
overallPassing = (avgMath + avgReading) / 2
#overallPassing = school_data_complete[(school_data_complete["math_score"] > 70) & (school_data_complete["reading_score"] > 70)].count()["school_name"]
#numberPassing = passingMath + passingReading
#Percentage passing
#percentOverall = (overallPassing / numberPassing) * 100
#percentPassingMath = (passingMath / totalStudents) * 100
#percentPassingReading = (passingReading / totalStudents) * 100
```

```python
#Percentage passing
percentPassingMath = (passingMath / totalStudents) * 100
percentPassingReading = (passingReading / totalStudents) * 100
```

```python
USDSummary = pd.DataFrame({"Total Schools":[totalSchools], 
                           "Total Students":[totalStudents], 
                           "Total Budget":[totalBudget], 
                           "Average Math Score":[avgMath],  
                           "Average Reading Score":[avgReading], 
                           "Passing Math":[passingMath], 
                           "Passing Reading":[passingReading], 
                           "Overall Passing Percent":[overallPassing], 
                           "Percent Passing Math":[percentPassingMath], 
                           "Percent Passing Reading":[percentPassingReading]})

USDSummary = USDSummary[["Total Schools", "Total Students", "Total Budget", "Average Math Score", 
                         "Average Reading Score", "Passing Math", "Passing Reading",  
                         "Percent Passing Math", "Percent Passing Reading", "Overall Passing Percent"]]

USDSummary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Summary

schoolType = school_data.set_index(["school_name"])["type"]
schools_groupby = school_data_complete.groupby(["school_name"])
totalStudentsSchool = schools_groupby.count()["Student ID"]
schoolBudget = schools_groupby.mean()['budget']
budgetStudent = schoolBudget / totalStudentsSchool
avgMathScore = schools_groupby.mean()["math_score"]
avgReadingScore = schools_groupby.mean()["reading_score"]
passingMathPercent = school_data_complete[school_data_complete["math_score"] > 70].groupby(["school_name"])
passingReadingPercent = school_data_complete[school_data_complete["reading_score"] > 70].groupby(["school_name"])
mathPer = passingMathPercent.count()["math_score"] / totalStudentsSchool *100
readingPer = passingReadingPercent.count()["reading_score"] / totalStudentsSchool * 100
overallPassing = (mathPer + readingPer)/2

```

```python
schoolSumm = pd.DataFrame({"School Type":schoolType, 
                           "Total Students":totalStudentsSchool, 
                           "Total Budget per School":schoolBudget, 
                           "Budget per Student":budgetStudent,  
                           "Average Math Score":avgMathScore, 
                           "Average Reading score":avgReadingScore, 
                           "Percent Passing Math":mathPer, 
                           "Percent Passing Reading":readingPer, 
                           "Overall Passing Rate":overallPassing, 
                           })

schoolSumm
```
